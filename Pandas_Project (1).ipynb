### Extract Day of the week

```python
# Lets create days of the week column
# Python assighns 0 to Monday and 6 to Sunday
# Example with the 1st value in 'Date' Column
absenteeism_data_RN['Date'][0].weekday()
```

```python
# Let's create a column with all the days of the week
# Let's create a function
def Day_of_the_Week(date_value):
    return date_value.weekday()
```

```python
absenteeism_data_RN['Day of the week'] = absenteeism_data_RN['Date'].apply(Day_of_the_Week)
absenteeism_data_RN.head()
```

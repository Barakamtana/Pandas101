```python
import os
!pip install gitpython
from git import Repo

def push_to_github(commit_message):
    # Change directory to the GitHub repository

    filepath ='/content/drive/MyDrive/Pandas101/Pandas101'

    os.chdir(filepath)

    # Commit and push changes
    repo = Repo(filepath)
    repo.git.add(".")
    repo.index.commit(commit_message)
    correction = repo.remote(name='correction')
    correction.push()

push_to_github('test commit')
```
